### Exploración enfocada

```python
colums_selected = ['ResponseId', 'Age', 'EdLevel', 'YearsCode', 'YearsCodePro', 'Country', 'CompTotal', 'LanguageHaveWorkedWith', 
                   'LanguageWantToWorkWith', 'LanguageAdmired', 'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith', 'DatabaseAdmired', 
                   'PlatformHaveWorkedWith', 'PlatformWantToWorkWith', 'PlatformAdmired', 'NEWCollabToolsHaveWorkedWith', 
                   'NEWCollabToolsWantToWorkWith', 'NEWCollabToolsAdmired', 'TechEndorse']
df_focus = df[colums_selected]
df_focus.head()
```

```python
df_focus.dtypes
```

```python
null_sum = pd.DataFrame(
    {
        'nulos': df_focus.isnull().sum(),
        'porcentaje': df_focus.isnull().mean() * 100
    }
).sort_values(by='porcentaje', ascending=False)
null_sum
```
